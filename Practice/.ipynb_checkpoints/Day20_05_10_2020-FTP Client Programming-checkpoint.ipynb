## FTP Client Programming

### cs.brown.edu ftp site

```python
from ftplib import FTP
```

```python
f = FTP('ftp.cs.brown.edu')
```

```python
f.login(user = 'anonymous', acct = 'iamidiot@gmail.com')
```

```python
f.dir()
```

```python
f.pwd()
```

```python
f.cwd('incoming')
```

```python
f.dir()
```

```python
f.storlines('STOR SS1.txt', open('sample.txt', 'rb'))
```

```python
f.dir()
```

```python
f.quit()
```
